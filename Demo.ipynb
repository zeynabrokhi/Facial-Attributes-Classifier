```python
import tensorflow
import cv2
import numpy as np
import time
import sys
```

```python
model_MobileNetV2 = tensorflow.keras.models.load_model("model_MobileNetV2.h5", compile=False)
```

```python
face_haar_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap=cv2.VideoCapture(0)

info1 = "No Data"
info2 = "No Data"
info3 = "No Data"
info4 = "No Data"
info5 = "No Data"
        
        
while True:
    ret,test_img=cap.read()# captures frame and returns boolean value and captured image
    if not ret:
        continue
    gray_img= cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)

    faces_detected = face_haar_cascade.detectMultiScale(gray_img, 1.32, 5)


    for (x,y,w,h) in faces_detected:
        cv2.rectangle(test_img,(x-20,y-40),(x+w+20,y+h+30),(200,200,0),thickness=7)
        roi_gray = gray_img[y-40:y+h+30, x-20:x+w+20]
        roi_color = test_img[y-40:y+h+30, x-20:x+w+20]
        
        roi_color = cv2.resize(roi_color, (64,64))
        roi_color = roi_color * 1./255
        roi_color = np.reshape(roi_color,[1,64,64,3])
        predicted = model_MobileNetV2.predict(roi_color)
        predicted = np.round(predicted[0,:],2)
        info1 = "Smiling: " + str(predicted[0])
        info2 = "Young: " + str(predicted[1])
#         info3 = "Bald: " + str(predicted[2])
        info3 = "EyeGlasses: " + str(predicted[2])
        info4 = "Straight_Hair: " + str(predicted[3])
        info5 = "Mustache: " + str(predicted[4])
        
    resized_img = cv2.resize(test_img, (1000, 700))
    cv2.putText(resized_img, info1, (int(20), int(80)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
    cv2.putText(resized_img, info2, (int(20), int(150)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
    cv2.putText(resized_img, info3, (int(20), int(220)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
    cv2.putText(resized_img, info4, (int(20), int(290)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
    cv2.putText(resized_img, info5, (int(20), int(360)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,255), 2)
    cv2.imshow('Facial analysis ',resized_img)


    if cv2.waitKey (1) & 0xFF == ord ('q'):#wait until 'q' key is pressed
        break

cap.release()
cv2.destroyAllWindows()
```
